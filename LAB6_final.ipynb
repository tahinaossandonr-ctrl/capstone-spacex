```python
# TASK 3: Calculate the distances between a launch site to its proximities
from folium.plugins import MousePosition

# Agregar MousePosition para leer coordenadas directo del mapa
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)
site_map.add_child(mouse_position)

# Funci√≥n Haversine (km)
from math import sin, cos, sqrt, atan2, radians
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0
    lat1, lon1, lat2, lon2 = map(radians, [lat1,lon1,lat2,lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1)*cos(lat2)*sin(dlon/2)**2
    c = 2*atan2(sqrt(a), sqrt(1-a))
    return R*c

# === PASOS ===
# 1) Elige un sitio (puedes cambiar el nombre por otro de launch_sites_df)
site_name = 'KSC LC-39A'  # ej.: 'CCAFS LC-40', 'VAFB SLC-4E'
site_row = launch_sites_df[launch_sites_df['Launch Site']==site_name].iloc[0]
site_lat, site_lon = float(site_row['Lat']), float(site_row['Long'])

# 2) Obt√©n coordenadas de proximidades con el MousePosition y p√©galas aqu√≠:
# (Ejemplos aproximados; reempl√°zalos por los tuyos del mapa)
coast_lat, coast_lon = 28.56367, -80.57163   # Costa
rail_lat,  rail_lon  = 28.57200, -80.58500   # Ferrocarril
high_lat,  high_lon  = 28.39285, -80.60564   # Autopista
city_lat,  city_lon  = 28.39220, -80.60770   # Ciudad cercana

# 3) Calcular distancias
d_coast = calculate_distance(site_lat, site_lon, coast_lat, coast_lon)
d_rail  = calculate_distance(site_lat, site_lon, rail_lat,  rail_lon)
d_high  = calculate_distance(site_lat, site_lon, high_lat,  high_lon)
d_city  = calculate_distance(site_lat, site_lon, city_lat,  city_lon)

d_coast, d_rail, d_high, d_city

```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# 4) Dibujar l√≠neas y etiquetas de distancia en el mapa
from folium.features import DivIcon

site_coord = (site_lat, site_lon)

def add_line_and_label(mapobj, target, color, label, km):
    folium.PolyLine([site_coord, target], color=color, weight=3,
                    tooltip=f"{label}: {km:.2f} km").add_to(mapobj)
    folium.Marker(
        target,
        icon=DivIcon(icon_size=(20,20), icon_anchor=(0,0),
                     html=f'<div style="font-size: 12px; color:{color};"><b>{km:.2f} KM</b></div>')
    ).add_to(mapobj)

add_line_and_label(site_map, (coast_lat, coast_lon), 'blue',   'Coast',   d_coast)
add_line_and_label(site_map, (rail_lat,  rail_lon),  'green',  'Railway', d_rail)
add_line_and_label(site_map, (high_lat,  high_lon),  'orange', 'Highway', d_high)
add_line_and_label(site_map, (city_lat,  city_lon),  'purple', 'City',    d_city)

from IPython.display import display, IFrame
display(site_map)
site_map.save("spacex_task3_distances.html")
display(IFrame("spacex_task3_distances.html", width="100%", height=520))
```

```python
# (Solo si NO tienes a√∫n site_lat y site_lon:)
# site_name = 'KSC LC-39A'  # o 'CCAFS LC-40', 'VAFB SLC-4E', etc.
# row = launch_sites_df[launch_sites_df['Launch Site']==site_name].iloc[0]
# site_lat, site_lon = float(row['Lat']), float(row['Long'])

from folium.features import DivIcon
import folium
from IPython.display import display

def add_proximity(name, poi_lat, poi_lon, color="#1f77b4"):
    """Dibuja l√≠nea sitio‚ÜíPOI y pone un marcador con la distancia (km)"""
    km = calculate_distance(site_lat, site_lon, poi_lat, poi_lon)
    # l√≠nea
    folium.PolyLine([(site_lat, site_lon), (poi_lat, poi_lon)],
                    color=color, weight=3,
                    tooltip=f"{name}: {km:.2f} km").add_to(site_map)
    # marcador con etiqueta de distancia
    folium.Marker(
        (poi_lat, poi_lon),
        tooltip=name,
        icon=DivIcon(icon_size=(20,20), icon_anchor=(0,0),
                     html=f'<div style="font-size:12px;color:{color};"><b>{km:.2f} KM</b></div>')
    ).add_to(site_map)
    return km

# === Pega aqu√≠ LAS COORDENADAS que tomes con MousePosition ===
# (Ejemplos; reemplaza por tus valores reales)
coast_lat, coast_lon = 28.56367, -80.57163   # Costa
city_lat,  city_lon  = 28.39220, -80.60770   # Ciudad cercana
rail_lat,  rail_lon  = 28.57200, -80.58500   # Ferrocarril
high_lat,  high_lon  = 28.39285, -80.60564   # Autopista

# Dibujar todo y devolver distancias
d_coast = add_proximity("Coastline", coast_lat, coast_lon, color="blue")
d_city  = add_proximity("City",      city_lat,  city_lon,  color="purple")
d_rail  = add_proximity("Railway",   rail_lat,  rail_lon,  color="green")
d_high  = add_proximity("Highway",   high_lat,  high_lon,  color="orange")

display(site_map)
print(f"Distancias (km) ‚Üí Costa: {d_coast:.2f} | Ciudad: {d_city:.2f} | Ferrocarril: {d_rail:.2f} | Autopista: {d_high:.2f}")

# (opcional) guarda el mapa para tu entrega
site_map.save("spacex_task3_distances.html")

```
